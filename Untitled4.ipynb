```python
# !wget http://prdownloads.sourceforge.net/ta-lib/ta-lib-0.4.0-src.tar.gz
# !tar -xzvf ta-lib-0.4.0-src.tar.gz
# %cd ta-lib
# !./configure --prefix=/usr
# !make
# !make install
# !pip install Ta-Lib
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
from datetime import date
from datetime import datetime
from talib import MACD, RSI
import warnings
warnings.filterwarnings("ignore")
import scipy.stats
from scipy.stats import norm
```

```python
index_data = pd.read_csv("/content/index_data.csv")
call_option_data = pd.read_csv("/content/call_option_data.csv")
put_option_data = pd.read_csv("/content/put_option_data.csv")
```

```python
# here we dropping NaN values
index_data.dropna(inplace=True)
call_option_data.dropna(inplace=True)
put_option_data.dropna(inplace=True)

# we are filling NaN values with mean
put_option_data.fillna(put_option_data.mean(), inplace=True)
call_option_data.fillna(call_option_data.mean(), inplace=True)
index_data.fillna(index_data.mean(), inplace=True)

# we are replacing '-' with NaN
put_option_data = put_option_data.replace('-', np.nan)
call_option_data = call_option_data.replace('-', np.nan)
index_data = index_data.replace('-', np.nan)

# we are dropping remaining NaN values
put_option_data.dropna(inplace=True)
call_option_data.dropna(inplace=True)

# stripping whitespaces from column names
put_option_data.columns = put_option_data.columns.str.strip()
call_option_data.columns = call_option_data.columns.str.strip()
index_data.columns = index_data.columns.str.strip()

# we are converting 'Date' column to datetime format
put_option_data['Date'] = pd.to_datetime(put_option_data['Date'])
call_option_data['Date'] = pd.to_datetime(call_option_data['Date'])
index_data['Date'] = pd.to_datetime(index_data['Date'])
```

```python
print("Summary Statistics for Put Option Data:")
print(put_option_data.describe())

print("\nSummary Statistics for Call Option Data:")
print(call_option_data.describe())

print("\nSummary Statistics for Index Data:")
print(index_data.describe())
```

```python
# calculating the MACD indicator for the closing prices,
# MACD is a trend-following momentum indicator that shows the relationship between two moving averages of the index's price.
# It consists of three components-  MACD line, signal line, and MACD histogram.
index_data['macd'], index_data['macd_signal'], _ = MACD(index_data['Close'])
index_data['rsi'] = RSI(index_data['Close'])
```

```python
macd_line, macd_signal_line, macd_histogram = MACD(index_data['Close'])
index_data['macd_histogram'] = macd_histogram
plt.figure(figsize=(12, 6))
plt.bar(index_data['Date'], index_data['macd_histogram'], color='green', width=0.5, label='MACD Histogram')


plt.axhline(y=0, color='gray', linestyle='--', linewidth=1)
plt.xlabel('Date')
plt.ylabel('MACD Histogram')
plt.title('MACD Histogram')
plt.legend()
plt.grid(True)
plt.show()
```

```python
index_option_correlation = index_data['Close'].corr(call_option_data['Close'])
print("Correlation between Index Movement and Option Data:", index_option_correlation)
```

```python
overbought_threshold = 70
oversold_threshold = 30
```

```python
index_data['Signal'] = 0
index_data['Signal'][(index_data['macd'] > index_data['macd_signal']) & (index_data['rsi'] < overbought_threshold)] = 1  # Buy signal
index_data['Signal'][(index_data['macd'] < index_data['macd_signal']) & (index_data['rsi'] > oversold_threshold)] = -1  # Sell signal
```

```python
index_data['PnL'] = index_data['Close'].pct_change() * index_data['Signal'].shift(1)
```

```python
index_data['Cumulative_Returns'] = (1 + index_data['PnL']).cumprod()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(index_data['Date'], index_data['Cumulative_Returns'], label='Strategy Returns')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.title('Trend-Following Trading Strategy Performance')
plt.legend()
plt.show()
```

```python
#  risk management parameters
risk_per_trade = 0.02  # Risk per trade (2% of total portfolio)
initial_portfolio_value = 100000  # Initial portfolio value


risk_per_position = initial_portfolio_value * risk_per_trade # position size based on risk per trade
position_size = risk_per_position / index_data['Close'][0]  #  first entry as initial entry


index_data['Signal'] = 0
index_data['Position'] = 0
index_data['Stop_Loss'] = 0

for i in range(1, len(index_data)):
    if index_data['macd'][i] > index_data['macd_signal'][i] and index_data['rsi'][i] < overbought_threshold:
        index_data['Signal'][i] = 1  # Buy signal
        index_data['Position'][i] = position_size
        index_data['Stop_Loss'][i] = index_data['Close'][i] - (index_data['Close'][i] * 0.02)  # Set stop loss at 2% below entry price
    elif index_data['macd'][i] < index_data['macd_signal'][i] and index_data['rsi'][i] > oversold_threshold:
        index_data['Signal'][i] = -1  # Sell signal
        index_data['Position'][i] = -position_size
        index_data['Stop_Loss'][i] = index_data['Close'][i] + (index_data['Close'][i] * 0.02)  # Set stop loss at 2% above entry price

# Backtesting using SLL
index_data['PnL'] = index_data['Position'].shift(1) * (index_data['Close'] - index_data['Close'].shift(1))
index_data['Cumulative_Returns'] = (1 + index_data['PnL']).cumprod()

for i in range(1, len(index_data)):
    if index_data['Position'][i-1] > 0 and index_data['Close'][i] < index_data['Stop_Loss'][i]:
        index_data['Position'][i] = 0
    elif index_data['Position'][i-1] < 0 and index_data['Close'][i] > index_data['Stop_Loss'][i]:
        index_data['Position'][i] = 0


index_data['PnL'] = index_data['Position'].shift(1) * (index_data['Close'] - index_data['Close'].shift(1)) # Re-calculate PnL
index_data['Cumulative_Returns'] = (1 + index_data['PnL']).cumprod() # recalculate and cumulative returns after applying stop-loss
plt.figure(figsize=(10, 6))
plt.plot(index_data['Date'], index_data['Cumulative_Returns'], label='Strategy Returns')
plt.xlabel('Date')
plt.ylabel('Cumulative Returns')
plt.title('Trend-Following Trading Strategy Performance with Risk Management')
plt.legend()
plt.show()
```

```python
# Define backtesting period
start_date = '2023-07-01'
end_date = '2023-12-31'

backtest_data = index_data[(index_data['Date'] >= start_date) & (index_data['Date'] <= end_date)]
#Filter data for backtesting period

# Initialize variables for performance metrics
total_profit = 0
total_loss = 0
winning_trades = 0
total_trades = 0

# Backtest the strategy
for i in range(1, len(backtest_data)):
    if backtest_data['Signal'][i] == 1:  # Buy signal
        entry_price = backtest_data['Open'][i + 1]  # Entry price on the next day's opening
        exit_price = backtest_data['Close'][i + 1]  # Exit price on the next day's closing
        profit_loss = exit_price - entry_price
        if profit_loss > 0:
            winning_trades += 1
        total_profit += max(0, profit_loss)
        total_loss += max(0, -profit_loss)
        total_trades += 1

# performance metrics
win_rate = (winning_trades / total_trades) * 100 if total_trades > 0 else 0
sharpe_ratio = (total_profit - total_loss) / np.std(backtest_data['PnL'])

print("Total Profit:", total_profit)
print("Total Loss:", total_loss)
print("Win Rate (%):", win_rate)
print("Sharpe Ratio:", sharpe_ratio)
```
